```python
from data.query.util import initializeSpotifyAPI
api = initializeSpotifyAPI()

from fuzzywuzzy import fuzz
from json import dump as jsondump
from pathlib import Path

allTracks = all_tracks

def songMatching(
    songToMatch: str, 
    artistToMatch: str,
    resultSong: str,
    resultArtists: list, 
    allowedRatio: int = 0.8) -> bool:
    
    blackList = ['Karaoke Version', 'Instrumental Version', '[Instrumental]', 'Emulation']
    
    if any(blacklisted.lower() in resultSong.lower() for blacklisted in blackList):
        return False
    
    if allowedRatio == 1:
        if songToMatch.lower() == resultSong.lower():
            # If song name matches check if the artist is found in artists
            # and if it is, this will be enough to add the song
            if 0.95 <= fuzz.token_set_ratio(artistToMatch, ' '.join((artist['name'] for artist in resultArtists))):
                return True
        return False
    
    if allowedRatio <= fuzz.token_set_ratio(songToMatch, resultSong):
        if allowedRatio <= fuzz.token_set_ratio(artistToMatch ,' '.join((artist['name'] for artist in resultArtists))):
            return True
        return False
    
    return False

def fetchSongsByNameFromSpotify(
    tracks: list,
    searchLimit: int = 1,
    matchingRatio:int = 0.8,
    useArtistInQuery: bool = True) -> list:
    
    found_songs = []
    no_match_indexes = []
    # This is used to remove duplicates
    madeQueries = set()
    duplicates = []
    for i, track in enumerate(tracks):
        billboard_song = track['song']
        billboard_artist = track['artist']
        
        query = billboard_song
        if useArtistInQuery:
            query = query + ' ' + billboard_artist
            
        if query not in madeQueries:
            madeQueries.add(query)
            no_match_indexes.append(i)
            #print("Querying: ", query)
            #print("Info: ", billboard_song, billboard:w
            _artist)
            for song in songQuery(api, query, searchLimit):
                #print("Result Song: ", song['name'])
                #print("Results Artist: ", ' '.join((artist['name'] for artist in song['artists'])))
                if songMatching(billboard_song, billboard_artist, song['name'], song['artists'], matchingRatio):
                    found_songs.append({
                        'spotifyData': song, 
                        'searchQuery': query, 
                        'matchingRatio': matchingRatio,
                        'originalDataSong': billboard_song,
                        'originalDataArtist': billboard_artist
                    })
                    no_match_indexes.pop()
                    break
        else:
            duplicates.append(i)
            
    return (found_songs, no_match_indexes, duplicates)

exactSongs, noMatchOne, dup = fetchSongsByNameFromSpotify(allTracks, 3, 1, False)
print("From %d exact matches: %d No matches: %d Duplicates: %d" % (len(allTracks), len(exactSongs), len(noMatchOne), len(dup)))                                                        
print("Saving exact matches")
      
# Save the exact matches
exactPath = Path('../data/datasets/spotify/exact_matches.json')
for i, parent in enumerate(exactPath.parents):
    if not parent.exists() and i < 2:
        parent.mkdir()
    else:
        break
with exactPath.open('w', encoding='utf8') as jf:
    jsondump(exactSongs, jf, ensure_ascii=False)
    
exactSongsWithArtistName, noMatchTwo, _ = fetchSongsByNameFromSpotify(
    (all_tracks[i] for i in noMatchOne), 
    3, 
    1, 
    True)
print("From %d exact matches: %d No matches with artist name in query: %d" % (len(noMatchOne), len(exactSongsWithArtistName), len(noMatchTwo)))                                                        
print("Saving exact matches with artist names in search")
      
# Save the exact matches
exactPathWithArtists = Path('../data/datasets/spotify/exact_matches_with_artist.json')
for i, parent in enumerate(exactPathWithArtists.parents):
    if not parent.exists() and i < 2:
        parent.mkdir()
    else:
        break
with exactPathWithArtists.open('w', encoding='utf8') as jf:
    jsondump(exactSongsWithArtistName, jf, ensure_ascii=False)
    
    
unexactSongs, noMatchThree, _ = fetchSongsByNameFromSpotify(
    (all_tracks[i] for i in noMatchOne), 
    3, 
    0.8,
    True)

print("From not exact songs %d with 80 match ratio found: %d no match: %d" % (len(noMatchTwo), len(unexactSongs), len(noMatchThree)))                                                        
print("Saving unexact matches")
# Save the unexact matches
unexactPath = Path('../data/datasets/spotify/unexact_matches.json')
for i, parent in enumerate(unexactPath.parents):
    if not parent.exists() and i < 2:
        parent.mkdir()
    else:
        break
with unexactPath.open('w', encoding='utf8') as jf:
    jsondump(unexactSongs, jf, ensure_ascii=False)
```

```python
from random import sample as rndsample

from data.query.util import initializeSpotifyAPI
api = initializeSpotifyAPI()

from data.query.billboard import getBillboardData
from data.query.spotify_api import getSpotifyDataFromBillboardSongs

billboardTracks = getBillboardData('../data/datasets/billboard/billboard-the-hot-100-songs.zip')
billboardSpotifyTrackData = getSpotifyDataFromBillboardSongs(api, billboardTracks, savePath='../data/datasets/spotify/query_results.json')

def trackBatcher(tracks, batchSize: int = 50):
    for i in range(0, len(tracks), batchSize):
        yield tracks[i:i+batchSize]

def getSpotifyAudioFeatures(api, tracks):
    
    allSpotifySongData = []
    for trackBatch in trackBatcher(tracks):
        audioFeatures = api.audio_features([track['spotifyData']['songID'] for track in trackBatch])
        
        for i, featuresResult in enumerate(audioFeatures):
            features: SpotifyFeatures = {}
            if featuresResult is not None:
                features: SpotifyFeatures = {
                    'timeSignature': featuresResult['time_signature'],
                    'durationMS': featuresResult['duration_ms'],
                    'key': featuresResult['key'],
                    'mode': featuresResult['mode'],
                    'acousticness': featuresResult['acousticness'], 
                    'danceability': featuresResult['danceability'], 
                    'energy': featuresResult['energy'], 
                    'instrumentalness': featuresResult['instrumentalness'], 
                    'liveness': featuresResult['liveness'], 
                    'loudness': featuresResult['loudness'], 
                    'speechiness': featuresResult['speechiness'], 
                    'valence': featuresResult['valence'], 
                    'tempo': featuresResult['tempo'] 
                }
            allSpotifySongData.append({'info': trackBatch[i], 'features': features})
            
    return allSpotifySongData

hitSongs = getSpotifyAudioFeatures(api, billboardSpotifyTrackData[:10])
print(hitSongs)

#class SpotifyRaw(TypedDict):
#    id: str


from data.types.spotify import SpotifySongInfo, SpotifyAlbum, SpotifyArtist

def parseSongInfo(songData, addAlbum: bool = True):
    print(songData.keys())
    songID: str = songData['id']
    songName: str = songData['name']
    artists: list[SpotifyArtist] = []
    for spotifyArtist in songData['artists']:
        artists.append({
            'name': spotifyArtist['name'], 
            'artistID': spotifyArtist['id']
            })
    
    if addAlbum:
        album: SpotifyAlbum = {
                'name': songData['album']['name'],
                'albumID': songData['album']['id'],
                'totalTracks': songData['album']['total_tracks'],
                'releaseDate': songData['album']['release_date']
                }

    info: SpotifySongInfo = {
            'name': songName,
            'songID': songID,
            'artists': artists,
            'album': album if addAlbum else None
            }
        
    return info

def fetchAlbumTracks(api, trackInfo, randomSelect: int = 5):
    # Loop album ids and fetch the track ids of tracks in every album
    allTracks = []
    for i, track in enumerate(trackInfo):
        album = track['spotifyData']['album']
        albumTracks = api.album_tracks(album['albumID'])['items']
        if len(albumTracks) > 1:
            # Take a random sample of tracks
            tracks = rndsample(albumTracks, randomSelect) if randomSelect > len(albumTracks) else rndsample(albumTracks, len(albumTracks))
            # Add ids of the random sample tracks to a list
            for albumTrack in albumTracks:
                if albumTrack['id'] != track['spotifyData']['songID']:
                    data = {'spotifyData': parseSongInfo(albumTrack, False)}
                    data['spotifyData']['album'] = album
                    data['searchQuery'] = "Shares album with " + track['spotifyData']['name']
                    allTracks.append(data)
            
        if i % 100 == 0:
            print("Album ", i+1)
        
    return allTracks

def getSongsWithAlbum(api, tracks, songNum):
    
    # Fetch the data for songs in a album
    randomSampleSize = 5
    print(tracks[0])
    songInfos = fetchAlbumTracks(api, tracks, randomSampleSize)
    # Fetch feature data for tracks
    return getSpotifyAudioFeatures(api, songInfos)
    

# Get not hits based on same album than hit song
notHitSongs = getSongsWithAlbum(api, billboardSpotifyTrackData[:10], 3)
print(len(notHitSongs))
print(notHitSongs)
```
